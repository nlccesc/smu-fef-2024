# Effect of Infrastructure investment in Laos on Tourism

Sources:

- Laos GDP Growth (%): https://data.worldbank.org/indicator/NY.GDP.MKTP.KD.ZG?end=2023&locations=LA&start=2014
- International tourism, number of arrivals (in millions): https://www.ceicdata.com/en/indicator/laos/visitor-arrivals
- Laos tourism Revenue (in millions): https://tradingeconomics.com/laos/tourism-revenues

```python
import matplotlib.pyplot as plt
```

```python
years = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]
gdp_growth = [7.6, 7.3, 7.0, 6.9, 6.2, 5.5, 0.5, 2.5, 2.7, 3.7]
tourism_arrivals = [4.159, 4.684, 4.239, 3.869, 4.186, 4.791, 0.886, 0.920, 1.294, 3.418]
tourism_revenue = [641.6, 724.8, 716.1, 648.1, 811, 934.7, 213.4, 0, 264.5, 927.8]
```

```python
fig, ax1 = plt.subplots(figsize=(12, 6))

line1, = ax1.plot(years, gdp_growth, marker='o', color='blue', linestyle='-', linewidth=2, markersize=6, label='GDP Growth (%)')
ax1.set_xlabel('Year')
ax1.set_ylabel('GDP Growth (%)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.grid(True, axis='y', linestyle='--', linewidth=0.5)

# Secondary y-axis for tourism arrivals
ax2 = ax1.twinx()
line2, = ax2.plot(years, tourism_arrivals, marker='o', color='green', linestyle='-', linewidth=2, markersize=6, label='Tourism Arrivals (in millions)')
ax2.set_ylabel('Tourism Arrivals (in millions)', color='green')
ax2.tick_params(axis='y', labelcolor='green')

# Adding a third y-axis for tourism revenue
ax3 = ax1.twinx()
ax3.spines['right'].set_position(('outward', 60))  # Offset the third axis
line3, = ax3.plot(years, tourism_revenue, marker='o', color='red', linestyle='-', linewidth=2, markersize=6, label='Tourism Revenue (in millions)')
ax3.set_ylabel('Tourism Revenue (in millions)', color='red')
ax3.tick_params(axis='y', labelcolor='red')

# Caption
fig.text(0.5, 0.92, 'Figure 8: Effects of Infrastucture investment in Laos on Tourism (2014-2022)', 
         ha='center', fontsize=12, fontweight='bold', color='white', backgroundcolor='#002f5f', 
         bbox=dict(facecolor='#002f5f', edgecolor='none', pad=5))

fig.text(0.10, 0.01, 'Source: https://github.com/nlccesc/smu-fef-2024/blob/main/laos.ipynb', 
         fontsize=8, fontstyle='italic', color='gray')

# Combine legends from all axes
lines = [line1, line2, line3]
labels = [line.get_label() for line in lines]
ax1.legend(lines, labels, loc='upper left')

plt.subplots_adjust(right=0.8)
plt.show()
```
