# Fiscal spending on tourism infrastructure

Sources:

- Tourist arrivals in the ASEAN region from 2013 to 2022(in million arrivals): https://www.statista.com/statistics/645643/tourist-arrivals-in-asean-region/#:~:text=In%202022%2C%20there%20were%20approximately,until%20the%20COVID%2D19%20pandemic.
- Gross Monthly Income From Employment (Including Employer CPF) Of Full-Time Employed Residents (median 50th percentile): https://tablebuilder.singstat.gov.sg/table/TS/M182981
Economy & Prices > Labour, Employment, Wages and Productivity > Wages and Income from Employment > Gross Monthly Income From Employment (Including Employer CPF) Of Full-Time Employed Residents
- Wages in Vietnam: https://www.ceicdata.com/en/indicator/vietnam/monthly-earnings

```python
import matplotlib.pyplot as plt
```

```python
years = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
tourist_arrivals = [105.08, 108.90, 115.57, 125.72, 135.17, 143.61, 26.16, 2.95, 43.15]
gross_income_sg = [3770, 3949, 4056, 4232, 4437, 4563, 4534, 1680, 5070]
monthly_earnings_vn = [225, 227, 228, 245, 258, 303, 305, 313, 260]
```

```python
fig, ax1 = plt.subplots(figsize=(12, 6))

bars = ax1.bar(years, tourist_arrivals, color='skyblue', alpha=0.7, label='Tourist Arrivals (in millions)')
ax1.set_xlabel('Year')
ax1.set_ylabel('Tourist Arrivals (in millions)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.grid(True, axis='y', linestyle='--', linewidth=0.5)

# Line plot for income data on the same plot but with a secondary y-axis
ax2 = ax1.twinx()
line1, = ax2.plot(years, gross_income_sg, marker='o', color='green', linestyle='-', linewidth=2, markersize=6, label='Gross Monthly Income in Singapore (SGD)')
line2, = ax2.plot(years, monthly_earnings_vn, marker='o', color='red', linestyle='-', linewidth=2, markersize=6, label='Monthly Earnings in Vietnam (USD)')
ax2.set_ylabel('Income/Earnings', color='gray')
ax2.tick_params(axis='y', labelcolor='gray')

# Caption
fig.text(0.5, 0.92, 'Figure 4: Tourist arrivals in the ASEAN region against wages in Singapore and Vietnam (2014-2022)', 
         ha='center', fontsize=12, fontweight='bold', color='white', backgroundcolor='#002f5f', 
         bbox=dict(facecolor='#002f5f', edgecolor='none', pad=5))

fig.text(0.10, 0.06, 'Source: https://github.com/nlccesc/smu-fef-2024/blob/main/fiscal_spending.ipynb', 
         fontsize=8, fontstyle='italic', color='gray')

lines = [bars, line1, line2]
labels = [line.get_label() for line in lines]
ax1.legend(lines, labels, loc='upper left')

plt.subplots_adjust(top=0.85, bottom=0.15)
plt.show()
```
