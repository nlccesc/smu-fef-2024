# Estimates of visitor numbers based on assumptions

- 2017: 1.2mill visitors (https://www.ncbi.nlm.nih.gov/pmc/articles/PMC8801303/)
- 'prior to 2018, 7000-8000 visitors a day' - so we estimate this with the following:

```python
import random

# Source: https://katewashere.com/2023/02/10/maya-bay-of-the-beach-life-after-the-closure/

# Number of days in a year
days_in_year = 365

# Daily visitors range
min_daily_visitors = 7000
max_daily_visitors = 8000

# Calculate random annual visitors for each year
random_annual_visitors_2014 = random.randint(min_daily_visitors, max_daily_visitors) * days_in_year
random_annual_visitors_2015 = random.randint(min_daily_visitors, max_daily_visitors) * days_in_year
random_annual_visitors_2016 = random.randint(min_daily_visitors, max_daily_visitors) * days_in_year

# Print the results
print(f"Random estimated annual visitors for 2014: {random_annual_visitors_2014}")
print(f"Random estimated annual visitors for 2015: {random_annual_visitors_2015}")
print(f"Random estimated annual visitors for 2016: {random_annual_visitors_2016}")
```

Following its opening in Jan 2022,
- "Maya Bay reopened to tourists, with new rules in place. Boats are forbidden from entering the front bay. Boats may drop off tourists at the back, and only 4,125 persons per day are allowed to visit."
- Source: https://katewashere.com/2023/02/10/maya-bay-of-the-beach-life-after-the-closure/
- Providing a rough estimate of visitors from 2022 to 2023:

```python
days_in_year = 365
max_daily_visitors = 4125

# Conservative estimate factor (e.g., 70% of minimum capacity, 90% of maximum capacity)
min_conservative_factor = 0.7
max_conservative_factor = 0.9

# Estimate annual visitors for each year
conservative_daily_visitors_2022 = int(max_daily_visitors * random.uniform(min_conservative_factor, max_conservative_factor))
conservative_daily_visitors_2023 = int(max_daily_visitors * random.uniform(min_conservative_factor, max_conservative_factor))

conservative_annual_visitors_2022 = conservative_daily_visitors_2022 * days_in_year
conservative_annual_visitors_2023 = conservative_daily_visitors_2023 * days_in_year

# Print the results for each year individually
print(f"Conservative estimated annual visitors for 2022: {conservative_annual_visitors_2022}")
print(f"Conservative estimated annual visitors for 2023: {conservative_annual_visitors_2023}")
```

- Post-closure and during COVID-19 (2018-2020): Significant increase in sightings, reflecting marine life recovery.
- Post-reopening (2022-2023): Some decline in sightings due to controlled reopening but still higher than pre-closure levels.

# Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Data

```python
years = np.array([2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023])
visitors = np.array([2646615, 2563030, 2611210, 1200000, 0, 0, 0, 0, 1279325, 1302320])
sightings = np.array([20, 25, 30, 35, 50, 80, 150, 170, 100, 90])

# Create DataFrame
data = {
    'Year': years,
    'Number of Visitors': visitors,
    'Marine Life Sightings': sightings
}
df = pd.DataFrame(data)
```

# Plotting

```python
fig, ax1 = plt.subplots(figsize=(12, 6))

# Number of Visitors on primary y-axis
line1 = ax1.plot(df['Year'], df['Number of Visitors'], 
                 color='skyblue', marker='o', label='Number of Visitors (million)')
ax1.set_xlabel('Year')
ax1.set_ylabel('Number of Visitors (millions)', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')

# Secondary y-axis for Marine Life Sightings
ax2 = ax1.twinx()
line2 = ax2.plot(df['Year'], df['Marine Life Sightings'], color='green', 
                 marker='o', linestyle='-', label='Marine Life Sightings')
ax2.set_ylabel('Marine Life Sightings', color='green')
ax2.tick_params(axis='y', labelcolor='green')

# Highlighting the closure and COVID-19 impact period
plt.axvspan(2017.5, 2021.5, color='gray', alpha=0.3, label='Closure and COVID-19 Impact')

# Combining legends from both axes
lines = line1 + line2
labels = [l.get_label() for l in lines]
ax1.legend(lines, labels, loc='upper left')

# Adding a caption and description
fig.text(0.5, 0.85, 'Figure 5: Relationship between Visitors and Marine Life Sightings at Maya Beach(2014-2023)', 
         ha='center', fontsize=10, fontweight='bold', color='white', backgroundcolor='#002f5f', 
         bbox=dict(facecolor='#002f5f', edgecolor='none', pad=5), fontfamily='Times New Roman')

# Positioning the source at the bottom left
fig.text(0.05, 0.075, 'https://jupyter.org/try-jupyter/lab/index.html?path=notebooks%2Fmaya.ipynb', 
         fontsize=8, fontstyle='italic')

plt.subplots_adjust(top=0.8, bottom=0.15)
plt.show()
```
