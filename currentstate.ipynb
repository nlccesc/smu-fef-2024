# Current state of tourism in ASEAN

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Data sourced from
- 'cambodia_gdp': Cambodia GDP 1960-2024. (n.d.). MacroTrends. https://www.macrotrends.net/global-metrics/countries/KHM/cambodia/gdp-gross-domestic-product#:~:text=Cambodia%20gdp%20for%202021%20was,a%2010.25%25%20increase%20from%202018.

- 'tourism_arrivals_cambodia': MINISTRY OF TOURISM, NATIONAL INSTITUTE OF TOURISM, & TOURISM STATISTICS DEPARTMENT. (2023). TOURISM STATISTICS REPORT. In MINISTRY OF TOURISM. https://www.nagacorp.com/eng/ir/tourism/tourism_statistics_202312.pdf

```python
# Data from above source
years = np.array([2014, 2015, 2016, 2017, 2018, 2019])
cambodia_gdp = np.array([16.70, 18.05, 20.02, 22.18, 24.57, 27.09]) # in billions
tourist_arrivals_cambodia= np.array([4502775, 4775231, 5011702, 5602157, 6201077, 6610592]) # dollars


# Create DataFrame
data = {
    'Year': years,
    'Cambodia GDP': cambodia_gdp,
    'Tourist arrivals in Cambodia': tourist_arrivals_cambodia,
}
df = pd.DataFrame(data)
```

```python
# Calculate percentage of maximum for GDP and Tourist Arrivals
df['GDP Percentage'] = df['Cambodia GDP'] / df['Cambodia GDP'].max() * 100
df['Tourist Arrivals Percentage'] = df['Tourist arrivals in Cambodia'] / df['Tourist arrivals in Cambodia'].max() * 100

# Create the figure and plot
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(df['Year'], df['GDP Percentage'], color='blue', marker='o', linestyle='-', linewidth=2, label='Cambodia GDP (%)')
ax.plot(df['Year'], df['Tourist Arrivals Percentage'], color='green', marker='s', linestyle='-', linewidth=2, label='Tourist Arrivals (%)')

# Set labels and title
ax.set_xlabel('Year')
ax.set_ylabel('Percentage of Maximum GDP and Tourist Arrivals')
# Display the legend
ax.legend()

# Adding a caption and description above the plot
fig.text(0.5, 0.85, 'Figure 2: Relationship between Cambodia GDP and Tourist Arrivals (pre-pandemic)', 
         ha='center', fontsize=10, fontweight='bold', color='white', backgroundcolor='#002f5f', 
         bbox=dict(facecolor='#002f5f', edgecolor='none', pad=5))

# Positioning the source at the bottom left
fig.text(0.01, 0.075, 'Source: https://jupyter.org/try-jupyter/lab/index.html?path=notebooks%2Fcurrentstate.ipynb', 
         fontsize=8, fontstyle='italic')

# Adjust layout to make room for the text
plt.subplots_adjust(top=0.8, bottom=0.15)

# Show the plot
plt.show()
```
