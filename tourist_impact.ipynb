# Impact of Tourism in Thailand

# Sources

- Gini Coefficient in Thailand: https://data.worldbank.org/indicator/SI.POV.GINI?end=2021&locations=TH&start=2015
- Thailand yearly tourist arrivals (millions): https://www.statista.com/statistics/994693/thailand-number-international-tourist-arrivals/
- Thailand CPI (Base B.E. 2562) (end year): https://www.thaibma.or.th/EN/CPI/CPIIndex.aspx
- Thailand house price index values (end year): https://tradingeconomics.com/thailand/housing-index
- Thailand inflation rate (end year): https://tradingeconomics.com/thailand/inflation-cpi


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Data for Thailand

```python
years = np.array([2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023])
gini_coefficient_th = np.array([0.36, 0.369, 0.364, 0.364, 0.349, 0.350, 0.349, 0.43, 0.43])
tourist_arrivals_th = np.array([29.9, 32.5, 35.59, 38.18, 39.8, 6.7, 0.43, 11.07, 28.15])
cpi_th = np.array([96.90, 97.99, 98.75, 99.10, 99.97, 99.70, 101.86, 107.86, 106.96])
house_price_index_th = np.array([127.9, 127.2, 133.8, 138.2, 140.2, 145.7, 149.2, 156.5, 160.9])
inflation_rate_th = np.array([-0.85, 1.13, 0.78, 0.36, 0.87, -0.27, 2.17, 5.89, -0.83])
```

# Dataframe

```python
data = {
    'Year': years,
    'Gini Coefficient TH': gini_coefficient_th,
    'Tourist Arrivals (Millions) TH': tourist_arrivals_th,
    'CPI TH': cpi_th,
    'House Price Index TH': house_price_index_th,
    'Inflation Rate (%) TH': inflation_rate_th
}
df = pd.DataFrame(data)
```

# Plot

```python
fig, ax1 = plt.subplots(figsize=(12, 8))

# Plot Gini Coefficient and Tourist Arrivals on the primary y-axis
ax1.plot(df['Year'], df['Gini Coefficient TH'], color='blue', marker='o', label='Gini Coefficient')
ax1.plot(df['Year'], df['Tourist Arrivals (Millions) TH'], color='orange', marker='o', label='Tourist Arrivals (Millions)')
ax1.set_xlabel('Year')
ax1.set_ylabel('Gini Coefficient / Tourist Arrivals (Millions)', color='black')
ax1.tick_params(axis='y', labelcolor='black')

# Secondary y-axis for CPI and House Price Index
ax2 = ax1.twinx()
ax2.plot(df['Year'], df['CPI TH'], color='green', marker='o', linestyle='-', label='CPI')
ax2.plot(df['Year'], df['House Price Index TH'], color='purple', marker='o', linestyle='-', label='House Price Index')
ax2.set_ylabel('CPI / House Price Index', color='black')
ax2.tick_params(axis='y', labelcolor='black')

# Third y-axis for Inflation Rate
ax3 = ax1.twinx()
ax3.spines['right'].set_position(('outward', 60))  # Offset the third axis
ax3.plot(df['Year'], df['Inflation Rate (%) TH'], color='red', marker='o', linestyle='-', label='Inflation Rate (%)')
ax3.set_ylabel('Inflation Rate (%)', color='red')
ax3.tick_params(axis='y', labelcolor='red')

# Combining legends from all axes
lines = ax1.get_lines() + ax2.get_lines() + ax3.get_lines()
labels = [line.get_label() for line in lines]
ax1.legend(lines, labels, loc='upper left')

# Adding a caption and description
fig.text(0.5, 0.85, 'Figure 6: Thailand Economic Indicators and Tourist Arrivals (2015-2023)', 
         ha='center', fontsize=10, fontweight='bold', color='white', backgroundcolor='#002f5f', 
         bbox=dict(facecolor='#002f5f', edgecolor='none', pad=5), fontfamily='Times New Roman')

# Positioning the source at the bottom left
fig.text(0.05, 0.075, 'https://github.com/nlccesc/smu-fef-2024/blob/main/tourist_impact.ipynb', 
         fontsize=8, fontstyle='italic')

# Adjust layout to make room for the text
plt.subplots_adjust(top=0.8, bottom=0.15)

# Show the plot
plt.show()
```
