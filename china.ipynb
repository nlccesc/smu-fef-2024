# SGD - RMB Exchange rate and its impact on tourism (miscellaneous)

Sources

- SGD to CNY historical rates: https://www.ofx.com/en-sg/forex-news/historical-exchange-rates/sgd/cny/
- International Visitor Arrivals By Inbound Tourism Markets (China) (in millions): https://tablebuilder.singstat.gov.sg/table/TS/M550001 - 
Industry > Tourism > International Visitor Arrivals (IVA) > International Visitor Arrivals By Inbound Tourism Markets

```python
import matplotlib.pyplot as plt
```

# Dataframe

```python
years = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]
sgd_to_cny = [4.571295, 4.811047, 4.894075, 4.904358, 5.065317, 5.002999, 4.801468, 4.879752, 5.26719]
visitor_arrivals_china = [15.231, 16.403, 17.425, 18.508, 19.116, 2.742, 0.330, 6.306, 13.602]
```

# Plot

```python
fig, ax1 = plt.subplots(figsize=(12, 6))

line1, = ax1.plot(years, sgd_to_cny, marker='o', color='blue', linestyle='-', linewidth=2, markersize=6, label='SGD to CNY Exchange Rate')
ax1.set_xlabel('Year')
ax1.set_ylabel('SGD to CNY Exchange Rate', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.grid(True, axis='y', linestyle='--', linewidth=0.5)

ax2 = ax1.twinx()
line2, = ax2.plot(years, visitor_arrivals_china, marker='o', color='green', linestyle='-', linewidth=2, markersize=6, label='Visitor Arrivals from China (in millions)')
ax2.set_ylabel('Visitor Arrivals from China (in millions)', color='green')
ax2.tick_params(axis='y', labelcolor='green')

# Caption
fig.text(0.5, 0.92, 'Figure 10: SGD to CNY Exchange Rate and Visitor Arrivals from China (2015-2023)', 
         ha='center', fontsize=12, fontweight='bold', color='white', backgroundcolor='#002f5f', 
         bbox=dict(facecolor='#002f5f', edgecolor='none', pad=5))


fig.text(0.10, 0.01, 'Source: https://github.com/nlccesc/smu-fef-2024/blob/main/china.ipynb', 
         fontsize=8, fontstyle='italic', color='gray')

# Combine legends from all axes
lines = [line1, line2]
labels = [line.get_label() for line in lines]
ax1.legend(lines, labels, loc='upper left')


plt.subplots_adjust(right=0.8)
plt.show()
```
