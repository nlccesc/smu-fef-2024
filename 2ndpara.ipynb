# Economic Impact of Increased Tourism in ASEAN

Import necessary libraries for generating the charts and graphs

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Data sourced from SingStat:
- 'tourism_receipts': Industry > Tourism > Tourism Receipts > Tourism Receipts By Major Components (Year-To-Date)
- 'hotel_room_revenue': Industry > Tourism > Hotel Industry > Hotel Statistics
- 'thai_visitors': Industry > Tourism > International Visitor Arrivals (IVA) > International Visitor Arrivals By Inbound Tourism Markets

```python
years = np.array([2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023])
tourism_receipts = np.array([59527, 53836, 63035, 67283, 67883, 68004, 17482, 4337, 29252, 65929])
hotel_room_revenue = np.array([3541784.50, 3507390.30, 3561442.60, 3715653.60, 3994617.40, 4195776.80, 1234008.10, 966248.60, 3243968.60, 4799284.30]) # thousand dollars
thai_visitors = np.array([506509, 516409, 546554, 531335, 545650, 528547, 63622, 4379, 283431, 450016])  # Data for Thai visitors

# Create DataFrame
data = {
    'Year': years,
    'Tourism Receipts': tourism_receipts,
    'Hotel Room Revenue': hotel_room_revenue,
    'Thai Visitors': thai_visitors
}
df = pd.DataFrame(data)
```

```python
fig, ax1 = plt.subplots(figsize=(12, 6))

# Tourism Receipts on primary y-axis
line1 = ax1.plot(df['Year'], df['Tourism Receipts'], color='blue', marker='o', label='Tourism Receipts (Million)')
ax1.set_xlabel('Year')
ax1.set_ylabel('Tourism Receipts (Million)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Secondary y-axis for Hotel Room Revenue
ax2 = ax1.twinx()
line2 = ax2.plot(df['Year'], df['Hotel Room Revenue'], color='orange', marker='o', label='Hotel Room Revenue (Thousand)')
ax2.set_ylabel('Hotel Room Revenue (Thousand)', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Third y-axis for Thai Visitors
ax3 = ax1.twinx()
ax3.spines["right"].set_position(("axes", 1.1))
line3 = ax3.plot(df['Year'], df['Thai Visitors'], color='green', marker='o', linestyle='-', label='Thai Visitors to SG')
ax3.set_ylabel('Thai Visitors', color='green')
ax3.tick_params(axis='y', labelcolor='green')

# Highlight COVID-19
plt.axvspan(2019.5, 2021.5, color='gray', alpha=0.3, label='COVID-19 Impact')

# Combining legends from all axes
lines = line1 + line2 + line3
labels = [l.get_label() for l in lines]
ax1.legend(lines, labels, loc='upper left')

# Adding a caption and description
fig.text(0.5, 0.85, 'Figure 3: Yearly Thailand tourists to Singapore and its impact on Hospitality and Retail', 
         ha='center', fontsize=10, fontweight='bold', color='white', backgroundcolor='#002f5f', 
         bbox=dict(facecolor='#002f5f', edgecolor='none', pad=5))

# Positioning the source at the bottom left
fig.text(0.05, 0.075, 'https://jupyter.org/try-jupyter/lab/index.html?path=notebooks%2F2ndpara.ipynb', 
         fontsize=8, fontstyle='italic')

plt.subplots_adjust(top=0.8, bottom=0.15)
plt.show()
```
